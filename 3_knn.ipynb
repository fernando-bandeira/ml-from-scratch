```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

bikes = np.random.multivariate_normal(
    mean=[1.8, 25],
    cov=[[0.1, 0.5], [0.5, 10]],
    size=30
)

cars = np.random.multivariate_normal(
    mean=[4.0, 160],
    cov=[[0.2, 2], [2, 50]],
    size=30
)

trucks = np.random.multivariate_normal(
    mean=[12.0, 100],
    cov=[[0.5, 3], [3, 30]],
    size=30
)

X = np.vstack((bikes, cars, trucks))
y = np.array(['bicycle']*30 + ['car']*30 + ['truck']*30)

plt.figure(figsize=(8,6))
plt.scatter(bikes[:,0], bikes[:,1], c="green", label="Bicycles")
plt.scatter(cars[:,0], cars[:,1], c="blue", label="Cars")
plt.scatter(trucks[:,0], trucks[:,1], c="red", label="Trucks")
plt.xlabel("Length (m)")
plt.ylabel("Max speed (km/h)")
plt.legend()
plt.grid(True)
plt.show()

X = X.tolist()
```

```python
def distance(a, b):
    return sum((x - y) ** 2 for x, y in zip(a, b)) ** 0.5

def knn(X, y, k, sample):
    distances_and_labels = [
        {'distance': distance(point, sample), 'label': label}
        for point, label in zip(X, y)
    ]
    distances_and_labels.sort(key=lambda p: p['distance'])
    neighbors = [neighbor['label'] for neighbor in distances_and_labels[:k]]
    return max(set(neighbors), key=neighbors.count)

print(knn(X, y, 5, [2, 20]))
print(knn(X, y, 5, [5, 150]))
print(knn(X, y, 5, [11, 90]))
```
