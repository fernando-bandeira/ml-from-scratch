```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 2.5 * x + 5 + np.random.normal(0, 2, size=x.shape)

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
m = 1
b = 1

def line_equation(m, b, x):
    return m * x + b

plt.scatter(x, y)
plt.plot(x, line_equation(m, b, x), color='red')
plt.show()
```

```python
alpha = 0.0001
m = 1
b = 1

def error_i(x_i, y_i):
    return m * x_i + b - y_i

for _ in range(1000):
    error = [error_i(x_i, y_i) for x_i, y_i in zip(x, y)]
    grad_m = 2 * sum([e_i * x_i for e_i, x_i in zip(error, x)])
    grad_b = 2 * sum(error)

    m = m - alpha * grad_m
    b = b - alpha * grad_b
plt.scatter(x, y)
plt.plot(x, line_equation(m, b, x), color='red')
plt.show()
```
